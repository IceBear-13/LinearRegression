```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("../data/house_prices.csv")
data
```

```python
# Plotting the data
fig, ax = plt.subplots(3, 2, figsize=(15, 15))

# Bedrooms vs Price
data.plot(x='Bedrooms', y='Price', kind='scatter', ax=ax[0, 0], title='Bedrooms vs Price')

# Bathrooms vs Price
data.plot(x='Bathrooms', y='Price', kind='scatter', ax=ax[0, 1], title='Bathrooms vs Price')

# SquareFeet vs Price
data.plot(x='SquareFeet', y='Price', kind='scatter', ax=ax[1, 0], title='SquareFeet vs Price')

# ProximityToCityCenter vs Price
data.plot(x='ProximityToCityCenter', y='Price', kind='scatter', ax=ax[1, 1], title='ProximityToCityCenter vs Price')

# Age vs Price
data.plot(x='Age', y='Price', kind='scatter', ax=ax[2, 0], title='Age vs Price')

# Hide the empty subplot
ax[2, 1].axis('off')

plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error 
```

```python
X = data[["Bedrooms", "Bathrooms", "SquareFeet", "ProximityToCityCenter", "Age"]]
y = data["Price"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)

# plot the pred and actual
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2)
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual vs Predicted Prices')
plt.show()
```

```python
example_house = [[4,1,1413,13,50]] 
predicted_price = model.predict(example_house)
print(f"Predicted price for the house: ${predicted_price[0]:.2f}")
```
